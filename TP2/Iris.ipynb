```python
1- Origine:
Le dataset Iris a été introduit par Ronald A. Fisher en 1936 dans son article "The Use of Multiple Measurements in Taxonomic Problems". Il est largement utilisé en statistiques, apprentissage automatique et analyse de données comme exemple classique de classification.

Variables: réparties en trois espèces de fleurs d'iris : Setosa, Versicolor, Virginica
Pour chaque fleur, les caractéristiques suivantes sont mesurées (en centimètres):
Longueur des sépales (sepal length)
Largeur des sépales (sepal width)
Longueur des pétales (petal length)
Largeur des pétales (petal width)
Espèce (species) : Classe cible (categorique) représentant l'espèce de la fleur.

Objectif d’étude:
L'objectif principal du dataset est de permettre l'analyse et la classification des fleurs d'iris en fonction de leurs caractéristiques physiques. Cela inclut :
Exploration des données : Examiner les relations entre les variables, identifier les patterns ou clusters.
Classification : Développer des modèles capables de prédire l'espèce d'une fleur donnée sur la base des mesures fournies.
Apprentissage supervisé : Utilisé comme exemple classique pour des algorithmes de classification tels que k-NN, SVM, Decision Trees ou Réseaux de Neurones.


```

```python
# question Num 2
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

# Charger le dataset Iris
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target
iris_df['species'] = iris_df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# 1. Matrice de corrélation
plt.figure(figsize=(8, 6))
corr_matrix = iris_df.drop('species', axis=1).corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matrice de Corrélation')
plt.show()

# 2. Nuage de points entre longueur de pétale et largeur de pétale
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris_df, x='petal length (cm)', y='petal width (cm)', hue='species', palette='Set2')
plt.title('Relation entre Longueur et Largeur des Pétales')
plt.show()

# 3. Boxplot pour chaque caractéristique en fonction de l'espèce
plt.figure(figsize=(12, 8))
sns.boxplot(data=iris_df, x='species', y='sepal length (cm)', palette='Set2')
plt.title('Boxplot : Longueur des Sépales en fonction de l\'Espèce')
plt.show()

# 4. Pairplot : visualisation de toutes les variables entre elles
sns.pairplot(iris_df, hue='species', palette='Set2')
plt.suptitle('Pairplot - Relations entre Variables', y=1.02)
plt.show()
```

```python
3- Observations remarquées dans les visualisations :

-Matrice de Corrélation (Heatmap)
Observation : La matrice de corrélation montre les relations entre toutes les paires de variables. Par exemple, la corrélation entre la longueur des pétales (petal length) et la largeur des pétales (petal width) est très élevée (proche de 1), ce qui indique que ces deux variables évoluent de manière similaire.
Interprétation : Une forte corrélation entre la longueur et la largeur des pétales suggère qu'un changement dans l'une de ces caractéristiques est probablement accompagné d'un changement similaire dans l'autre. Cela peut être utilisé pour prédire l'une de ces caractéristiques à partir de l'autre.

-Nuage de Points (Longueur et Largeur des Pétales)
Observation : Le nuage de points montre une séparation claire entre les trois espèces (setosa, versicolor, virginica) en fonction de la longueur des pétales et de la largeur des pétales. Les points de l'espèce setosa forment un cluster distinct avec des valeurs de longueur et de largeur plus faibles, tandis que versicolor et virginica se chevauchent un peu mais restent relativement séparées.
Interprétation : Les espèces setosa sont facilement distinguées de versicolor et virginica en raison de la différence marquée de taille des pétales. Les autres espèces se chevauchent plus, ce qui pourrait rendre la classification plus difficile pour un modèle d'apprentissage automatique. Cette visualisation souligne l'importance de la longueur des pétales et largeur des pétales comme caractéristiques discriminantes.

-Boxplot (Longueur des Sépales par Espèce)
Observation : Le boxplot montre que la longueur des sépales est plus petite pour setosa (avec une médiane proche de 5 cm) par rapport aux espèces versicolor et virginica, qui ont des longueurs de sépales plus grandes (médians autour de 5.5 cm à 6 cm).
Interprétation : Cela montre une différence claire dans les distributions des longueurs de sépales entre les espèces. Setosa présente une distribution plus concentrée autour de la médiane, tandis que versicolor et virginica montrent une plus grande variation. Cette information pourrait être utilisée pour distinguer les espèces dans une tâche de classification.

-Pairplot (Relations entre toutes les Variables)
Observation : Le pairplot permet de voir que certaines paires de variables, comme longueur des pétales et largeur des pétales, présentent une séparation claire entre les espèces. D'autres paires, comme longueur des sépales et largeur des sépales, présentent plus de chevauchements entre les espèces.
Interprétation : Certaines variables (comme la longueur et la largeur des pétales) sont de bons séparateurs pour les différentes espèces, tandis que d'autres (comme la largeur et la longueur des sépales) peuvent être moins discriminantes, surtout entre versicolor et virginica. Le pairplot nous aide à voir quels attributs sont les plus utiles pour une tâche de classification.

En outre, La longueur et largeur des pétales sont les meilleures variables pour séparer les espèces, surtout setosa des autres.
Les sépales ne sont pas aussi discriminants, surtout pour différencier versicolor et virginica.
La corrélation entre la longueur et la largeur des pétales est forte, ce qui suggère qu'elles sont liées de manière significative.
```

```python
# question num
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# Charger le dataset Iris
iris = load_iris()
X = iris.data
y = iris.target

# Diviser les données en ensembles d'entraînement et de test (70% pour l'entraînement et 30% pour le test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardiser les données (important pour KNN)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)



# Créer le modèle de régression logistique
log_reg = LogisticRegression(max_iter=200)

# Entraîner le modèle
log_reg.fit(X_train, y_train)

# Faire des prédictions
y_pred_log_reg = log_reg.predict(X_test)

# Calculer la précision
log_reg_accuracy = accuracy_score(y_test, y_pred_log_reg)
print(f'Précision de la régression logistique : {log_reg_accuracy:.4f}')


# Tester différentes valeurs de k pour KNN
k_values = range(2, 21)
knn_accuracies = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, y_train)
    y_pred_knn = knn.predict(X_test_scaled)
    knn_accuracies.append(accuracy_score(y_test, y_pred_knn))

# Afficher les résultats
plt.plot(k_values, knn_accuracies, marker='o')
plt.title('Précision de KNN en fonction de k')
plt.xlabel('k (nombre de voisins)')
plt.ylabel('Précision')
plt.xticks(k_values)
plt.grid(True)
plt.show()

# Trouver le meilleur k
best_k = k_values[np.argmax(knn_accuracies)]
print(f'Le meilleur k est : {best_k}')

# Créer le modèle KNN avec le meilleur k trouvé
knn_best = KNeighborsClassifier(n_neighbors=best_k)

# Entraîner le modèle
knn_best.fit(X_train_scaled, y_train)

# Faire des prédictions
y_pred_knn_best = knn_best.predict(X_test_scaled)

# Calculer la précision
knn_best_accuracy = accuracy_score(y_test, y_pred_knn_best)
print(f'Précision du modèle KNN avec k={best_k} : {knn_best_accuracy:.4f}')
```

```python
5- Avec les précisions qu'on a obtenues :

Régression Logistique : 1.0000
KNN avec k=3 : 1.0000
Interprétation des résultats
Les deux modèles ont atteint une précision parfaite de 1.0000, ce qui signifie qu'ils ont correctement classé toutes les observations dans l'ensemble de test. Cela suggère que :

Régression Logistique : Le modèle de régression logistique a parfaitement classé toutes les observations, ce qui est un excellent résultat, surtout dans le cas de la classification avec un dataset comme Iris.

KNN avec k=3 : Le modèle KNN avec k=3 a également atteint une précision parfaite. Le fait que le k=3 soit le meilleur choix indique que ce nombre de voisins est optimal pour ce dataset particulier, permettant au modèle de bien généraliser les classes.


Les deux modèles ont atteint la même précision parfaite. Cependant, cela ne signifie pas nécessairement que l'un est intrinsèquement meilleur que l'autre. Voici quelques considérations supplémentaires :

Complexité du modèle :
-Régression Logistique : Ce modèle est linéaire et relativement simple. Il offre une grande interprétabilité et fonctionne bien sur des données avec des relations linéaires entre les caractéristiques et les classes.
-KNN : Bien que KNN ait également atteint la même précision, il est souvent plus coûteux en termes de calcul, surtout pour de grandes quantités de données. Cela est dû à la nécessité de calculer la distance entre chaque point de test et tous les points d’entraînement à chaque prédiction.

Robustesse :
-La régression logistique, bien qu'efficace pour des problèmes linéaires, pourrait ne pas être aussi performante si les relations entre les variables sont complexes ou non linéaires.
-KNN peut mieux capturer des relations complexes, mais il est sensible au bruit dans les données et peut être plus lent à prédire avec de grandes quantités de données.


Puisque les deux modèles ont la même précision, la décision sur le modèle le plus performant pourrait se baser sur d'autres critères tels que :
Vitesse d'entraînement et de prédiction : Si le temps d'entraînement et de prédiction est un facteur important, la régression logistique peut être un meilleur choix, car elle est plus rapide et plus efficace, en particulier pour de grands ensembles de données.
Interprétabilité : La régression logistique est généralement plus facile à interpréter, ce qui peut être un avantage dans des situations où vous devez expliquer pourquoi une certaine prédiction a été faite.
Sensibilité au bruit : KNN peut être sensible au bruit et aux données aberrantes, ce qui pourrait affecter ses performances dans des cas réels.
En résumé, bien que les deux modèles aient montré des performances parfaites sur ce dataset, la régression logistique est souvent préférable en raison de sa simplicité, de sa rapidité et de son interprétabilité. KNN est aussi performant, mais il peut être plus coûteux en termes de calculs, surtout pour des jeux de données plus grands.
```
